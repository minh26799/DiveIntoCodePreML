```python
import matplotlib.pyplot as plt
TODAYGRAIN = 1
def compute_sorori_shinzaemon(day=100):
    list_n_grains = [TODAYGRAIN]
    list_total_grains = [0]
    for i in range(0, day - 1):
        list_n_grains.append(list_n_grains[i] * 2)
        list_total_grains.append(sum(list_n_grains))
    return list_n_grains, list_total_grains

# 1 pound = 7000 grains = 0.45359237kg
# As I found on the Internet, a person normally use 0.2468 kg of rice a day
RICEUSING = 0.2468
AGRAIN = 0.45359237/7000
def living(number_of_grians, number_of_people):
    daylyeating = int(number_of_people) * RICEUSING
    kgofrice = int(number_of_grians)*AGRAIN
    return kgofrice//daylyeating

days = input("Please, input day value: ")
# I choose the input value is fifty days.
list_n_grains, list_total_grains = compute_sorori_shinzaemon(day=int(days))
# In this assignment. I assume that there are one thousand people in a simulation society. 
NUMBER_OF_PEOPLE = 1000
living_day = living(list_total_grains[len(list_total_grains)-1],NUMBER_OF_PEOPLE) 
print("The Society with {} people can live {: .0f} days with the quantity of rice received after {} days.".format(NUMBER_OF_PEOPLE,living_day,days))
```
