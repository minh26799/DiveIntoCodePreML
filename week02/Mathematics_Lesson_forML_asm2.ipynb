```python
import numpy as np
import matplotlib.pyplot as plt
csv_path = "mtfuji_data.csv" # Specify the file name (path)

np.set_printoptions(suppress=True) # Set the prohibition of exponential notation
fuji = np.loadtxt(csv_path, delimiter=",", skiprows=1)
print(len(fuji))
```

```python
plt.title("Visualization of Mt. Fuji")
plt.xlabel("position")
plt.ylabel("elevation(m)")
plt.plot(fuji[:,0],fuji[:,3])
plt.show()
```

```python
def compute_gradient(ndarray):
    xarray = ndarray[:,0][1:] - ndarray[:,0][:-1]
    yarray = ndarray[:,1][1:] - ndarray[:,1][:-1]
    ndarrayA = np.array([])
    for each in tuple(zip(xarray,yarray)):
        ndarrayA = np.append(ndarrayA, each[1]/each[0])
    gradient = np.array([])
    gradient = ndarrayA
    return gradient
ndarray = np.array(tuple(zip(fuji[:,0],fuji[:,3])))
gradient = compute_gradient(ndarray)
tst = np.array(tuple(zip(fuji[:,0],gradient)))
print(gradient)
```

```python
prob3 = np.array([0])
for i in range(299):
    prob3 = np.append(prob3, prob3[i]+0.2*gradient[i])
plt.title("Mt.Fuji Visualization")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(prob3, color="red")
plt.show()
```

```python
def goingdown(number):
    downward = np.array(ndarray[:number][::-1])
    return downward
goingdownA = goingdown(136)
goingdownB = goingdown(145)
print(goingdownA)
plt.title("Mt.Fuji Going Down Visualization")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(goingdownA[:,1], color="red")
plt.show()
plt.title("Mt.Fuji Going Down Visualization")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(goingdownB[:,1], color="red")
plt.show()
```
