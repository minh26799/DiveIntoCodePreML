```python
import numpy as np
import matplotlib.pyplot as plt
#Problem 01
matrix = [[1.0, 0.8], [0.8, 1.0]]
datastored = np.array(np.random.multivariate_normal([-3,0],matrix,500)).reshape(500,2)
print("500 random numbers: {}".format(datastored))
```

```python
#Problem 02
plt.title("visualization")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.scatter(datastored[:,0],datastored[:,1])
plt.show()
```

```python
#Problem 03
ax1 = plt.subplot(131)
ax1.hist(datastored[:,0])
ax1.set_xlim(min(datastored[:,0]),max(datastored[:,0]))
ax1.title.set_text('histogram of x1')
ax1.set_xlabel('x1')
ax1.set_ylabel('frequency')
ax2 = plt.subplot(133)
ax2.hist(datastored[:,1])
ax2.set_xlim(min(datastored[:,1]),max(datastored[:,1]))
ax2.title.set_text('histogram of x2')
ax2.set_xlabel('x2')
ax2.set_ylabel('frequency')
plt.show()
```

```python
#Problem 04
datastored_prob4 = np.array(np.random.multivariate_normal([0,-3],matrix,500)).reshape(500,2)
plt.title("scatter")
plt.xlabel("x2")
plt.ylabel("x1")
plt.scatter(datastored[:,0],datastored[:,1])
plt.scatter(datastored_prob4[:,0],datastored_prob4[:,1])
plt.show()
```

```python
#Problem 05
# By np.vstack
combined = np.vstack((datastored,datastored_prob4)) 
# By np.concatenate
concatenated = np.concatenate((datastored,datastored_prob4))
```

```python
#Problem 06
newArray = np.array([])
for element in combined:
    if element in datastored:
        element = np.append(element, [0.])
    else:
        element = np.append(element, [1.])
    newArray = np.append(newArray, element)
newArray = newArray.reshape(1000,3)
print("The ndarray after labeling: \n {}".format(newArray))
```
